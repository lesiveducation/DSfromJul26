В результате преобразования у вас должен получиться столбец, содержащий информацию о том, сколько месяцев проработал соискатель.
Выполните преобразование, ответьте на контрольные вопросы и удалите столбец **"Опыт работы"** из таблицы.


```python
def exp_mnth (text):

    if pd.isna(text) or not isinstance(text, str):
        return np.nan
    
    # Приводим к нижнему регистру для сравнения
    text_lower = text.lower().strip()
       
    # Если текст слишком короткий или не содержит цифр - скорее всего нет данных
    if len(text) < 3 or not re.search(r'\d+', text):
        return np.nan
    
    years = 0
    months = 0
    
    # Поиск лет (год/года/лет)
    years_match = re.search(r'(\d+)\s*(?:год|года|лет)', text)
    if years_match:
        years = int(years_match.group(1))
    
    # Поиск месяцев (месяц/месяца/месяцев)
    months_match = re.search(r'(\d+)\s*(?:месяц|месяца|месяцев)', text)
    if months_match:
        months = int(months_match.group(1))
    
    # Если не нашли ни лет, ни месяцев - возвращаем NaN
    if years == 0 and months == 0:
        return np.nan
    
    total_months = years * 12 + months
    return total_months

hhdata['Опыт мес'] = hhdata['Опыт работы'].apply(exp_mnth)



exp_median = hhdata['Опыт мес'].median()
print(exp_median)

hhdata.head(5)
```

```python
MILLION_CITIES = {
    'Новосибирск', 'Екатеринбург', 'Нижний Новгород', 'Казань',
    'Челябинск', 'Омск', 'Самара', 'Ростов-на-Дону', 'Уфа',
    'Красноярск', 'Пермь', 'Воронеж', 'Волгоград'
}

# =====================================================
# 2. ФУНКЦИЯ ПАРСИНГА БЕЗ ИСКЛЮЧЕНИЙ
# =====================================================

def extract_city_no_exceptions(text):
    """
    Извлекает город из строки БЕЗ учета слов-исключений.
    Используется для анализа ошибок.
    """
    if pd.isna(text) or not isinstance(text, str):
        return np.nan
    
    text = str(text).strip()
    
    # Приоритет 1: Москва
    if 'Москва' in text:
        return 'Москва'
    
    # Приоритет 2: Санкт-Петербург
    if 'Санкт-Петербург' in text or 'СПб' in text:
        return 'Санкт-Петербург'
    
    # Приоритет 3: Города-миллионники
    for city in MILLION_CITIES:
        if city in text:
            return city
    
    # Приоритет 4: Первое слово с заглавной буквы (БЕЗ ИСКЛЮЧЕНИЙ!)
    first_word_match = re.search(r'^([А-Я][а-я]+(?:-[А-Я][а-я]+)?)', text)
    if first_word_match:
        return first_word_match.group(1)
    
    return 'другие'

def analyze_city_errors(df, location_column='Город, переезд, командировки'):
    """
    Анализирует результаты парсинга городов без исключений
    и выявляет потенциальные ошибки.
    """
    # Создаем копию данных
    df_analysis = hhdata.copy()
    
    # Парсим города БЕЗ исключений
    df_analysis['city_raw'] = df_analysis['Город, переезд, командировки'].apply(extract_city_no_exceptions)
    
    # 1. ТОП-20 самых частых "городов"
    print("\n" + "=" * 80)
    print("📊 ТОП-20 САМЫХ ЧАСТЫХ 'ГОРОДОВ' (БЕЗ ИСКЛЮЧЕНИЙ)")
    print("=" * 80)
    
    city_counts = df_analysis['city_raw'].value_counts().head(20)
    print(city_counts.to_string())
    
    # 2. Выявляем подозрительные слова (потенциальные ошибки)
    # Список слов, которые точно не являются городами
    known_non_cities = {
        'Готов', 'Не', 'Нет', 'Без', 'С', 'По', 
        'На', 'В', 'Из', 'От', 'До', 'При',
        'Опыт', 'Работа', 'Стаж', 'Год', 'Месяц',
        'Более', 'Менее', 'Около', 'Примерно',
        'Высшее', 'Среднее', 'Начальное'
    }
    
    # Находим потенциальные ошибки
    potential_errors = []
    for city, count in city_counts.items():
        if city in known_non_cities:
            potential_errors.append((city, count, 'явно не город'))
        elif len(city) <= 2:
            potential_errors.append((city, count, 'слишком короткое слово'))
        elif city.isdigit():
            potential_errors.append((city, count, 'цифры'))
        elif city in ['И', 'А', 'Но', 'За']:
            potential_errors.append((city, count, 'предлог/союз'))
    
    # 3. Выводим подозрительные города
    print("\n" + "=" * 80)
    print("⚠️ ПОДОЗРИТЕЛЬНЫЕ 'ГОРОДА' (ВЕРОЯТНЫЕ ОШИБКИ)")
    print("=" * 80)
    
    if potential_errors:
        df_errors = pd.DataFrame(potential_errors, columns=['Слово', 'Частота', 'Причина'])
        print(df_errors.to_string(index=False))
    else:
        print("✅ Подозрительных слов не найдено")
    
    # 4. Показываем примеры строк с подозрительными словами
    print("\n" + "=" * 80)
    print("📝 ПРИМЕРЫ СТРОК С ПОДОЗРИТЕЛЬНЫМИ СЛОВАМИ")
    print("=" * 80)
    
    suspicious_words = [item[0] for item in potential_errors if item[0] in known_non_cities]
    
    for word in suspicious_words[:5]:  # показываем первые 5
        examples = df_analysis[df_analysis['city_raw'] == word][location_column].head(3)
        print(f"\nСлово: '{word}' (встречается {len(df_analysis[df_analysis['city_raw'] == word])} раз)")
        for example in examples:
            print(f"  → {example[:80]}...")
    
    # 5. Рекомендации по добавлению в список исключений
    print("\n" + "=" * 80)
    print("💡 РЕКОМЕНДАЦИИ ПО ДОБАВЛЕНИЮ В СПИСОК ИСКЛЮЧЕНИЙ")
    print("=" * 80)
    
    new_exceptions = []
    for word, count, reason in potential_errors:
        if word in known_non_cities:
            new_exceptions.append(word)
    
    if new_exceptions:
        print("Рекомендуется добавить следующие слова в список исключений:")
        print(f"EXCLUDED_WORDS = {set(new_exceptions)}")
        
        # Проверяем, какие уже есть в текущем списке
        current_exceptions = {'Готов', 'Не', 'Нет', 'Без', 'С', 'По'}
        new_words = set(new_exceptions) - current_exceptions
        if new_words:
            print(f"\nНовые слова для добавления: {new_words}")
        else:
            print("\nВсе подозрительные слова уже есть в списке исключений ✅")
    else:
        print("Новых слов для исключения не найдено")
    
    return df_analysis

# =====================================================
# 4. ДЕТАЛЬНЫЙ АНАЛИЗ ОШИБОК
# =====================================================

def detailed_error_analysis(df, location_column='location'):
    """
    Более детальный анализ ошибок с группировкой по типам.
    """
    df_analysis = df.copy()
    df_analysis['city_raw'] = df_analysis[location_column].apply(extract_city_no_exceptions)
    
        # Категоризация результатов
    def categorize_city(city):
        if pd.isna(city):
            return 'пустое'
        if city in ['Москва', 'Санкт-Петербург']:
            return 'ключевой'
        if city in MILLION_CITIES:
            return 'миллионник'
        if city in ['Готов', 'Не', 'Нет', 'Без', 'С', 'По', 'На', 'В', 'Из']:
            return 'служебное_слово'
        if len(city) <= 2:
            return 'короткое_слово'
        if city.isdigit():
            return 'цифры'
        return 'другой_город'
    
    df_analysis['category'] = df_analysis['city_raw'].apply(categorize_city)
    
    print("\n" + "=" * 80)
    print("📊 КАТЕГОРИЗАЦИЯ РЕЗУЛЬТАТОВ ПАРСИНГА")
    print("=" * 80)
    
    category_stats = df_analysis['category'].value_counts()
    print(category_stats.to_string())
    
    # Детали по каждой категории
    categories = {
        'служебное_слово': 'Служебные слова (ошибки!)',
        'короткое_слово': 'Слова короче 3 букв (вероятные ошибки)',
        'цифры': 'Цифры (точно ошибки)',
        'другой_город': 'Другие города (вероятно корректные)',
        'ключевой': 'Москва/СПб (корректно)',
        'миллионник': 'Города-миллионники (корректно)',
        'пустое': 'Пустые значения'
    }
    
    print("\n" + "=" * 80)
    print("🔍 ДЕТАЛИ ПО КАТЕГОРИЯМ")
    print("=" * 80)
    
    for category, description in categories.items():
        if category in category_stats.index:
            count = category_stats[category]
            examples = df_analysis[df_analysis['category'] == category]['city_raw'].head(5).tolist()
            print(f"\n{description}: {count} записей")
            print(f"  Примеры: {', '.join(map(str, examples))}")
    
    return df_analysis
    
hhdata_analyzed = analyze_city_errors(hhdata)
#df_detailed = detailed_error_analysis(hhdata)
```

Сейчас признаки представляют собой набор категорий желаемой занятости (полная занятость, частичная занятость, проектная работа, волонтерство, стажировка) и желаемого графика работы (полный день, сменный график, гибкий график, удаленная работа, вахтовый метод).
На сайте hh.ru соискатель может указывать различные комбинации данных категорий, например:
* полная занятость, частичная занятость
* частичная занятость, проектная работа, волонтерство
* полный день, удаленная работа
* вахтовый метод, гибкий график, удаленная работа, полная занятость

Такой вариант признаков имеет множество различных комбинаций, а значит множество уникальных значений, что мешает анализу. Нужно это исправить!

Давайте создадим признаки-мигалки для каждой категории: если категория присутствует в списке желаемых соискателем, то в столбце на месте строки рассматриваемого соискателя ставится True, иначе - False.

Такой метод преобразования категориальных признаков называется One Hot Encoding и его схема представлена на рисунке ниже:
<img src=https://raw.githubusercontent.com/AndreyRysistov/DatasetsForPandas/main/ohe.jpg>
Выполните данное преобразование для признаков "Занятость" и "График", ответьте на контрольные вопросы, после чего удалите их из таблицы

```python
full = lambda x: (
    False if pd.isna(x) or not isinstance(x, str) else True if 'полная' in x else False
)

part = lambda x: (
    False if pd.isna(x) or not isinstance(x, str) else True if 'частичная' in x else False
)

project = lambda x: (
    False if pd.isna(x) or not isinstance(x, str) else True if 'проектная' in x else False
)

intern = lambda x: (
    False if pd.isna(x) or not isinstance(x, str) else True if 'стажировка' in x else False
)

vol = lambda x: (
    False if pd.isna(x) or not isinstance(x, str) else True if 'волонт' in x else False
)

vacht = lambda x: (
    False if pd.isna(x) or not isinstance(x, str) else True if 'вахт' in x else False
)
flex  = lambda x: (
    False if pd.isna(x) or not isinstance(x, str) else True if 'гибк' in x else False
)

hhdata['Полная занятость'] = hhdata['Занятость'].apply(full)
hhdata['Частичная занятость'] = hhdata['Занятость'].apply(part)
hhdata['Проетная работа'] = hhdata['Занятость'].apply(project)
hhdata['Стажировка'] = hhdata['Занятость'].apply(intern)
hhdata['Волонтерство'] = hhdata['Занятость'].apply(vol)
hhdata['Вахта'] = hhdata['График'].apply(vacht)
hhdata['Гибкий график'] = hhdata['График'].apply(flex)

hhdata.head(5)
```

```python
proj_vol = (hhdata['Проетная работа'] == True) & (hhdata['Волонтерство'] == True)
count_proj_vol = proj_vol.sum()
print(count_proj_vol)

vacht_fl = (hhdata['Вахта'] == True) & (hhdata['Гибкий график'] == True)
count_vacht_fl = vacht_fl.sum()
print(count_vacht_fl)
```
